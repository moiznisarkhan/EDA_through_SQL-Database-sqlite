```python
import sqlite3 as sql
import pandas as pd
```

```python
conn = sql.connect('database.sqlite')
conn
```

```python
pd.read_sql_query("select * from salaries", conn)
```

```python
def sql_query(q):
    return pd.read_sql_query(q, conn)
```

## In

look into the data for desired results

```python
q = '''select employeename, jobtitle from salaries
    where jobtitle = "CAPTAIN III (POLICE DEPARTMENT)"
    or 
    jobtitle = "WIRE ROPE CABLE MAINTENANCE MECHANIC"
    '''

sql_query(q)
```

```python
q = '''select employeename, jobtitle from salaries
    where jobtitle in ("CAPTAIN III (POLICE DEPARTMENT)", "WIRE ROPE CABLE MAINTENANCE MECHANIC", "Custodian")
'''

sql_query(q)
```

```python
q = '''select employeename, jobtitle from salaries
    where jobtitle not in ("CAPTAIN III (POLICE DEPARTMENT)", "WIRE ROPE CABLE MAINTENANCE MECHANIC", "Custodian")
'''

sql_query(q)
```

case is used inplace of if, when as elif & else as else

```python
# Categorizing Employees Based on overtimepay > 100000 'overtime worker' else 'regular worker'
q = '''select employeename, overtimepay,
    case
    when overtimepay > 100000 then "overtime worker"
    else "regular worker"
    end as "Work Category"
    from salaries '''
sql_query(q)
```

```python
# Categorizing Employees Based on BasePay > 100000 'high paying job', between 50k to 99999 'medium paying job', else 'low paying job'
q = '''select employeename, basepay,
    case 
    when basepay > 100000 then "High paying job"
    when basepay between 50000 and 99999 then "Medium paying job"
    else "low paying job"
    end as Paying_Category
    from salaries'''
sql_query(q)
```

## Update

update table_name set value where...

update is used to modify data...
On live server we got a button to execute and the modified data is saved permenantly on servers.
Whereas on offline files we use ".execute" for execution and to save the modified data permenantly on files we use ".commit".

```python
q='''update salaries 
    set totalpay = '9999999'
    where employeename = 'NATHANIEL FORD'
'''

conn.execute(q)
```

```python
q='''select * from salaries'''

sql_query(q)
```

```python
q='''update salaries 
    set totalpay = '9999999'
    where employeename = 'NATHANIEL FORD'
'''

conn.execute(q)
conn.commit()
```

```python
q='''select * from salaries'''

sql_query(q)
```
