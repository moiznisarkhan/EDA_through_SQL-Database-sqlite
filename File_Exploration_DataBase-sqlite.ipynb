```python
#function

def sql_query(q):
    return pd.read_sql_query(q, conn)
```

## Ascending/Descending.

```python
# For Ascending or Descending => (order by) is used, bydefault ascending or you may write it as (order by asc) & for descending (order by desc) is used.
```

```python
q = 'select basepay, totalpay from salaries'

sql_query(q)
```

```python
q = '''select basepay, totalpay, otherpay from salaries order by 2'''

sql_query(q)
```

```python
q = "select basepay, totalpay, otherpay from salaries order by 2 desc"

sql_query(q)
```

```python
q = '''select basepay, totalpay, otherpay from salaries order by totalpay'''

sql_query(q)
```

```python
q = '''select basepay, totalpay, otherpay from salaries order by totalpay desc'''

sql_query(q)
```

```python
q = 'select employeename, jobtitle from salaries where employeename = "Not provided"'

sql_query(q)
```

```python
q = 'select employeename, jobtitle from salaries where employeename = "Not Provided"'

sql_query(q)
```

## Limit.

```python
q = "select employeename, totalpay from salaries order by 2 limit 10"

sql_query(q)
```

```python
q = "select employeename, totalpay from salaries order by 2 desc limit 10"

sql_query(q)
```

```python
q = "select employeename, totalpay from salaries where employeename != 'Not provided' and employeename != 'Not Provided'"

sql_query(q)
```

```python
q = "select employeename, totalpay from salaries where employeename != 'Not provided' or employeename != 'Not Provided'"

sql_query(q)
```

```python
q = "select employeename, totalpay from salaries where employeename is not 'Not provided' and employeename is not 'Not Provided'"

sql_query(q)
```
